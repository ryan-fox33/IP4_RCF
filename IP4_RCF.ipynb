```python
cs=pd.read_csv('car-sales.csv')
cs.head()
```

```python
plt.plot(cs['Price'])
plt.xlabel('Row')
plt.ylabel('Price ($)')
plt.show()
```

```python
plt.plot(cs['Odometer (KM)'])
plt.xlabel('Row')
plt.ylabel('Odometer (KM)')
plt.show()
```

### Odometer works because it is an integer while price does not work because it is an object

```python
cs.head()
```

```python
columns=['Make','Colour','Odometer (KM)','Doors','Price']
for col in columns:
    cs[col] = cs[col].map(lambda x: str(x).replace('$',''))
    
    
for col in columns:
    cs[col] = cs[col].map(lambda x: str(x).replace(',',''))
```

```python
cs.head()
```

```python
cs['Price']=cs['Price'].astype('int64')
```

```python
a1= np.array(cs['Price'])
out_sum = np.cumsum(a1,dtype='int64') 
cs.insert(5, "Total Sales", out_sum, True)
cs.head()
```

```python
sale_date=pd.date_range(start='2-12-2023',periods=len(cs),freq='D')

cs.insert(6, "Sale Date", sale_date, True)
cs.head()
```

```python
plt.plot(cs['Price'],cs['Total Sales'])
plt.xlabel('Price')
plt.ylabel('Total Sales')
plt.title('Price vs. Total Sales Line Plot')
plt.show()
```

```python
cs['Odometer (KM)']=cs['Odometer (KM)'].astype('int64')
```

```python
plt.scatter(cs['Odometer (KM)'],cs['Price'])
plt.xlabel('Odometer (KM)')
plt.ylabel('Price')
plt.title('Odometer (KM) vs. Price Scatter Plot')
plt.show()
```

```python
plt.bar(cs['Make'],cs['Odometer (KM)'])
plt.xlabel('Make')
plt.ylabel('Odometer (KM)')
plt.title('Make vs. Odometer (KM) Bar Plot')
plt.show()
```

```python
plt.hist(cs['Make'])
plt.xlabel('Make')
plt.title('Histogram of Car Make')
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))

ax1=fig.add_subplot(221)
ax1.set_title('Price vs. Total Sales Line Plot')

ax1.plot(cs['Price'],cs['Total Sales'])
ax1.set_xlabel('Price')
ax1.set_ylabel('Total Sales')

ax2=fig.add_subplot(222)
ax2.set_title('Odometer (KM) vs. Price Scatter Plot')

ax2.scatter(cs['Odometer (KM)'],cs['Price'])
ax2.set_xlabel('Odometer (KM)')
ax2.set_ylabel('Price')

ax3=fig.add_subplot(223)
ax3.set_title('Make vs. Odometer (KM) Bar Plot')

ax3.bar(cs['Make'],cs['Odometer (KM)'])
ax3.set_xlabel('Make')
ax3.set_ylabel('Odometer (KM)')

ax4=fig.add_subplot(224)
ax4.set_title('Histogram of Car Make')

ax4.hist(cs['Make'])
ax4.set_xlabel('Make')

plt.show()
```
